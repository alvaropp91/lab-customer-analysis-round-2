```python
#Check and deal with NaN values.

nulls_df = pd.DataFrame(round(data.isna().sum()/len(data),4)*100)
nulls_df
nulls_df = nulls_df.reset_index()
nulls_df
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
# Delete the columns with NaN values with a % between 5.70 and 5.80

data = data.dropna(subset=['State', 'Response', 'Number of Open Complaints', 'Months Since Last Claim', 'Vehicle Class', 'Vehicle Size'])
data
```

```python
# Change NaN values in the column Vehicle Type for "Unknown"
data['Vehicle Type'] = data['Vehicle Type'].fillna("")
```

```python
# Replace the name of a value for another one (from "U" to "Unknown")
data['Vehicle Type'] = data['Vehicle Type'].replace(['U'], 'Unknown')
```

```python
data
```

```python
data['Month'] = pd.DatetimeIndex(data['Effective To Date']).month
```

```python
data
```

```python
data['Month'] = pd.to_datetime(data['Effective To Date'], format='%m/%d/%y').dt.strftime('%B')
data
```

```python
# Filter the months of January, February and March in the DF
data[data['Month'].isin(["January", "February", "March"])]
```

```python
data
```

```python
# BONUS: Put all the previously mentioned data transformations into a function

def action(data):
    data = data.rename(columns={'EmploymentStatus':'Employment Status'})
    data = data.dropna(subset=['State', 'Response', 'Number of Open Complaints', 'Months Since Last Claim', 'Vehicle Class', 'Vehicle Size'])
    data['Vehicle Type'] = data['Vehicle Type'].fillna("")
    data['Vehicle Type'] = data['Vehicle Type'].replace(['U'], 'Unknown')
    data['Month'] = pd.DatetimeIndex(data['Effective To Date']).month
    data['Month'] = pd.to_datetime(data['Effective To Date'], format='%m/%d/%y').dt.strftime('%B')
    data[data['Month'].isin(["January", "February", "March"])]
    return data
```
